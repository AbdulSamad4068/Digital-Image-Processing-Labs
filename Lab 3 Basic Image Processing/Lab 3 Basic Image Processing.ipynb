## Task 1 
Image Binarization using a predefined global threshold

```python
import numpy as np
from PIL import Image

%pylab inline
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_b1 = Image.open('B1.png').convert('L')
img_b2 = Image.open('B2.jpg').convert('L')
img_b3 = Image.open('B3.jpg').convert('L')
```

```python
def binarize(img, name, threshold):
    arr = np.asarray(img).copy()
    arr[arr < threshold] = 0    
    arr[arr >= threshold] = 255 
    imageArray = Image.fromarray(arr)
    
    #Save image in this naming format
    imageArray.save('result' + '_' + name + '.png')
```

```python
# Threshold = 140
binarize(img_b1, 'img_b1', threshold = 140)
result_b1 = Image.open('result_img_b1.png')
result_b1.show()
```

```python
# Threshold = 130
binarize(img_b2, 'img_b2', threshold = 130)
result_b2 = Image.open('result_img_b2.png')
result_b2.show()
```

```python
# Threshold = 185
binarize(img_b3, 'img_b3', threshold=185)
result_b3 = Image.open('result_img_b3.png')
result_b3.show()
```

## Task 2
Create Intensity Histogram from a Greyscale image

```python
def cal_intensity_values(img):
    img_arr = np.asarray(img).copy()
    intensity_vals = np.zeros(256)

    for pixel in np.nditer(img_arr):
        intensity_vals[pixel] = intensity_vals[pixel] +1

    return intensity_vals
```

```python
def plot_bargraph(intensity_arr, bins):
    plt.bar(range(bins),intensity_arr)
    # The following commands add labels to our figure.
    plt.xlabel('Intensity Values')
    plt.ylabel('Instensity values count')
    plt.title('Vertical Bar chart of intensity values')
    plt.show()
```

### Bar graph of intensity values of B1.png

```python
intensity_arr_img1 = cal_intensity_values(img_b1)
plot_bargraph(intensity_arr_img1, 256)
```

### Bar graph of intensity values of B2.jpg

```python
intensity_arr_img2 = cal_intensity_values(img_b2)
plot_bargraph(intensity_arr_img2, 256)
```

### Bar graph of intensity values of B3.png

```python
intensity_arr_img3 = cal_intensity_values(img_b3)
plot_bargraph(intensity_arr_img3, 256)
```

## Task 3
Apply Recursive XY-cut algorithm

```python
img_orig_xy_cut = Image.open('XY-cuts-cropped.png').convert('L')
img_arr = np.asarray(img_orig_xy_cut).copy()
img_arr.shape
```

```python
# Threshold = 140
binarize(img_orig_xy_cut, 'XY-cuts-cropped-binarized', threshold = 140)
binarized_img = Image.open('result_XY-cuts-cropped-binarized.png')
binarized_img.show()
```

```python
binarized_img_arr = np.asarray(binarized_img).copy()
```

```python
intensity_arr = np.zeros(binarized_img_arr.shape[0])

for row in range(len(binarized_img_arr)):
    for column in range(len(binarized_img_arr[row])):
        if (binarized_img_arr[row][column] == 0):
            intensity_arr[row] = intensity_arr[row] + 1
```

```python
plot_bargraph(intensity_arr, img_arr.shape[0])
```

```python
img_xy_cut = np.asarray(binarized_img).copy()
#print(img_xy_cut[125])

img_xy_cut[130] = 0
img_xy_cut[270] = 0
img_xy_cut[390] = 0
img_xy_cut[525] = 0
imageArray = img_xy_cut

#print('\n\n\n\n',imageArray)

imageArray = Image.fromarray(imageArray)
imageArray.save('result' + '_' + 'img_xy_cut' + '.png')
#print(imageArray)
```
